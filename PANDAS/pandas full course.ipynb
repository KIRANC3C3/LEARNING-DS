### how to handle missing data(2 ways)

1)remove missing data------DROPNA()

2)fill with default values-----FILLNA()

>FOR variablename.dropna() orginal data is not affected

>for variablename.dropna(inplace=True)  orginal data is affected

>for variablename.fillna() orginal data is not affected

>for variablename.fillna(inplace=True)  orginal data is affected



## how to handle with duplicate data

*we can remove duplicates 

variablename.drop_duplicates()




```python
#data filtering using conditions
mydata.loc[(mydata['Miles']<50)]
```

```python
# AND CONDITION
mydata.loc[(mydata['Miles']<50) & (mydata['Age']<50)]
```

```python
# OR CONDITION (any one condition should satisfy)
mydata.loc[(mydata['Miles']<50) | (mydata['Age']<50)]
```
