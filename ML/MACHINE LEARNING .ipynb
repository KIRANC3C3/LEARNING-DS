# 3) gradient decent

> mean square error
 
 

 >Gradient descent is an algorithm that finds best fit line for given training data set
    
 > In gradient descent first we approximate some value  
     
     example:
     
     >if the total exam marks attempted by tinku is 50
     
     >when i used to ask about paper attemption then he says guess
     
     >then i start with a value and it will goes near to 0 he will reveal?
     
     >now in this case { new value=old value-step size}
     
    > in machine learning step size is learning rate*slope
    

>The global minimum is the least value of a function while a local minimum is the least value of a function in a certain neighbourhood

> at each point we need to calculate slope

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    rate = 0.001
    n = len(x)
    for i in range(10):
        y_predicted = m_curr * x + b_curr
        md = -(2/n)*sum(x*(y-y_predicted))
        yd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd
       # print('m {},b {}'.format(m_curr,b_curr))
        plt.plot(x,y_predicted)
        plt.plot(x,y,color='red')
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```

# correct way

```python
# we need use scale
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
scaler.fit(a[['Income($)']])
a['Income($)']=scaler.transform(a[['Income($)']])
a
```

```python
plt.scatter(a0.Age,a0['Income($)'],color='green')
plt.scatter(a1.Age,a1['Income($)'],color='red')
plt.scatter(a2.Age,a2['Income($)'],color='black')

plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(a[['Age','Income($)']])
y_predicted
```

```python
a['cluster']=y_predicted
a
```

```python
a0=a[a.cluster==0]
a1=a[a.cluster==1]
a2=a[a.cluster==2]

plt.scatter(a0.Age,a0['Income($)'],color='green')
plt.scatter(a1.Age,a1['Income($)'],color='red')
plt.scatter(a2.Age,a2['Income($)'],color='black')

plt.xlabel('Age')
plt.ylabel('Income($)')
```

# centroid

```python
cent=km.cluster_centers_
cent
```

```python
a0=a[a.cluster==0]
a1=a[a.cluster==1]
a2=a[a.cluster==2]

plt.scatter(a0.Age,a0['Income($)'],color='green',label='(Income($)')
plt.scatter(a1.Age,a1['Income($)'],color='red',label='(Income($)')
plt.scatter(a2.Age,a2['Income($)'],color='black',label='(Income($)')
plt.scatter(cent[:,0],cent[:,1],color='purple',marker='+',label='centroid')
plt.xlabel('Age')
plt.ylabel('Income($)')
# cent[:,0]  ...... means (: )all rows from cent and (0) is first column
#cent[:,1].........means (: )all rows from cent and (1) is first column
```

# elbow method

```python
k_rang=range(1,10)
sse=[]
for k in k_rang:
    km=KMeans(n_clusters=k)
    km.fit(a[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
k_rang
```
