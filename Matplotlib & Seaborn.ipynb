# seaborn

```python
#setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import data 
a=pd.read_csv('CardioGoodFitness-1.csv')
a
```

```python
print(sns.get_dataset_names())
```

# distribution plot


```python
sns.distplot(a['Miles'])
```

# JOINT PLOT

```python
sns.jointplot(x='Age',y='Miles',data=a,kind='kde')
```

# kde plot

```python
sns.kdeplot(a['Usage'])
```

# pair plots

```python
sns.pairplot(a)
```

# rug plot

```python
sns.rugplot(a['Income'])
# IT PLOTS SINGLE STICK IN THE GRAPH
```

# STYLING

```python
sns.set_style('dark')
sns.jointplot(x='Age',y='Miles',data=a,kind='reg')
```

# categorical plot

## barplot

```python
sns.barplot(x='Gender',y='Miles',data=a,estimator=np.mean)
```

# count plot

```python
sns.countplot(x='Gender',data=a)
```

# box plot

```python
sns.boxplot(x='Product',y='Miles',hue='Gender',data=a)
```

# violin plot

```python
sns.violinplot(x='Product',y='Miles',hue='Gender',data=a)
```

# strip plot

```python
sns.stripplot(x='Product',y='Miles',hue='Gender',data=a)
```

# swarm plot

```python
sns.swarmplot(x='Product',y='Miles',hue='Gender',data=a)
# palettes are used to to change colour
# hue is used for categorical  data 
```

# heatmaps

```python
corr=a.corr()
corr
sns.heatmap(corr,annot=True,cmap='Blues')
```
